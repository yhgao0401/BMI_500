```python
import numpy as np
import pandas as pd
import random
```

```python
sex=['male','female']
race=['Am. Ind/AK Native','Asian','Black or Afr. Amer','White']

#age 18-30
#height 145-220cm
#weight 40-200kg

df_id=[]
df_sex=[]
df_race=[]
df_age=[]
df_height=[]
df_weight=[]

random.seed(42)

for i in range(1000):
    df_id.append(i)
    df_sex.append(random.choice(sex))
    df_race.append(random.choice(race))
    df_age.append(random.randint(18, 30))
    df_height.append(random.randint(145, 220))
    df_weight.append(random.randint(40, 200))

df=pd.DataFrame({'id':df_id, 'sex': df_sex, 'race': df_race, 'age': df_age, 'height (cm)': df_height, 'weight (kg)': df_weight})
df
```

```python
df['race'].value_counts()
```

```python
df['bmi']=df['weight (kg)']/(df['height (cm)']/100)**2
df=df[df['bmi']>=18.5]
df
```

```python
df['bmi'].describe()
```

```python
# df['level_general']=np.where(df['bmi']<25, 'healthy',
#                             np.where(df['bmi']<30, 'overweight',
#                                     np.where(df['bmi']<35, 'obesity 1',
#                                             np.where(df['bmi']<40, 'obesity 2','obesity 3'))))
# df['level_alternative']=np.where(df['race']!='Asian', df['level_general'],
#                                 np.where(df['bmi']<23, 'healthy',
#                                          np.where(df['bmi']<25,'overweight',
#                                                 np.where(df['bmi']<30, 'obesity 1', 'obesity 2'))))


df['level_general']=np.where(df['bmi']<25, 'Healthy',
                            np.where(df['bmi']<30, 'Overweight', 'Obese'))
df['level_alternative']=np.where(df['race']!='Asian', df['level_general'],
                                np.where(df['bmi']<23, 'Healthy',
                                         np.where(df['bmi']<25,'Overweight','Obese')))
df
                                        
```

```python
df_asian_differ=df[df['level_general']!=df['level_alternative']]
df_asian_differ.shape
```

```python
df_all_same=df[df['level_general']==df['level_alternative']].sample(n=67,random_state=42)
df_all_same
```

```python
df_synthetic=pd.concat([df_all_same,df_asian_differ])
df_synthetic
```

```python
df_synthetic['race'].value_counts()
```

```python
with open('synthetic.txt','w') as f:
    id=1
    for i in df_synthetic.itertuples():
        # print(i[1])
        s=f'patient {id}: sex-{i[2]}, race-{i[3]}, age-{i[4]}, height-{i[5]}cm, weight-{i[6]}kg\n'
        # print(s)
        f.write(s)
        id+=1
```

```python
df_synthetic.to_excel('HW9_data.xlsx', index=False)
```
